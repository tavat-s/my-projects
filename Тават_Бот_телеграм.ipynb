```python
import requests
from bs4 import BeautifulSoup
import random

url = "https://buklya.com/top-100-knig"
response = requests.get(url)
soup = BeautifulSoup(response.text)
books = soup.find_all('span', class_='top_100_book_title')
authors = soup.find_all('span', class_='top_100_book_author')
description = soup.find_all('p')


bot = telebot.TeleBot(TOKEN)
# Функция, обрабатывающая команду /start
@bot.message_handler(commands=['start'])
def start(message):
    bot.send_message(message.chat.id, 'Здравствуйте! Отправьте любую цифру для того, чтобы получить рекомендацию по книге ')

# Получение сообщений от пользователя
@bot.message_handler(content_types=["text"])
def handle_text(message):
    if message.text.lower().isdigit():
            number_book = random.randint(0, len(books) - 1)
            books_1 = books[number_book].text
            authors_1 = authors[number_book].text
            description_1 = description[number_book].text
            new_book = f'"{books_1}" {authors_1}. \n Описание книги:  {description_1}'
            bot.send_message(message.chat.id, new_book)
    else:
        bot.send_message(message.chat.id, 'Введите любую цифру для того, чтобы получить рекомендацию по книге ')

# Запускаем бота
bot.polling(none_stop=True, interval=0)
```
